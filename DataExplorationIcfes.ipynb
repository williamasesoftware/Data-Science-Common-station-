```python
# Packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
num_columns = len(df.columns)
print(f"The dataframe has {num_columns} columns.")
```

```python
column_types = df.dtypes
print("Column types:")
print(column_types)
```

```python
column_name="ESTU_NACIONALIDAD"
unique_values = df[column_name].unique()
print(f"The {column_name} column has {len(unique_values)} unique values:")
print(unique_values)
```
